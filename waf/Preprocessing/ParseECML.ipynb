```python
import xml.etree.ElementTree as ET
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set consistent styling for all plots
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 11
```

```python
data = []
```

```python
root = ET.parse('C:\\Users\\mps\\Desktop\\AegisVault\\AegisVault\\waf\\Datasets\\learning_dataset.xml').getroot()
```

```python
for child in root:
    sample = {}
    #print(child.tag)
    #print(child.attrib)
    sample['id'] = child.attrib['id']
    for subchild in child:
        #rint(subchild.tag)
        for subsub in subchild:
        #   print(subsub.tag)
            sample[subsub.tag] = subsub.text
            #if subsub.tag == '{http://www.example.org/ECMLPKDD}attackIntervall':
            #    print (subsub.text)
        #print()
    #break
    data.append(sample)
```

```python
final_data = []
```

```python
import random
import math

for index, row in df.iterrows():
    sample = {}
    sample['type'] = row['{http://www.example.org/ECMLPKDD}type']
    if row['{http://www.example.org/ECMLPKDD}type'] == 'Valid':
        new = True
        while(new):
            c = random.randint(0, 3)
            if c==0:
                sample['request'] = row['{http://www.example.org/ECMLPKDD}uri']
            elif c==1:
                sample['request'] = row['{http://www.example.org/ECMLPKDD}query']
            elif c==2:
                sample['request'] = row['{http://www.example.org/ECMLPKDD}body']
            elif c==3:
                headers = row['{http://www.example.org/ECMLPKDD}headers'].split('\n')
                whereis = random.sample(['Cookie', 'User-Agent', 'Accept-Language', 'Accept-Encoding'], 1)[0]
                for h in headers:
                    if h.startswith(whereis):
                        sample['request'] = h[len(whereis)+2:].strip()
                        break
            if ('request' in sample) and (sample['request'] != None):
                if type(sample['request']) == str:
                    if sample['request'] != 'nan':
                        new = False
                else:
                    if not math.isnan(sample['request']):
                        new = False
    else:
        sample['interval'] = row['{http://www.example.org/ECMLPKDD}attackIntervall']
        if row['{http://www.example.org/ECMLPKDD}attackIntervall'].startswith('uri'):
            sample['request'] = row['{http://www.example.org/ECMLPKDD}uri']
        elif row['{http://www.example.org/ECMLPKDD}attackIntervall'].startswith('query'):
            sample['request'] = row['{http://www.example.org/ECMLPKDD}query']
        elif row['{http://www.example.org/ECMLPKDD}attackIntervall'].startswith('body'):
            sample['request'] = row['{http://www.example.org/ECMLPKDD}body']
        elif row['{http://www.example.org/ECMLPKDD}attackIntervall'].startswith('headers'):
            whereis = row['{http://www.example.org/ECMLPKDD}attackIntervall'][8:]
            idx = whereis.find(':')
            whereis = whereis[:idx]
            headers = row['{http://www.example.org/ECMLPKDD}headers'].split('\n')
            for h in headers:
                if h.startswith(whereis):
                    sample['request'] = h[len(whereis)+2:].strip()
                    break
    final_data.append(sample)
```

```python
final_df = pd.DataFrame(final_data)
```

```python
# Display first 10 rows
final_df.head(10)
```

```python
final_df1 = final_df[final_df.type == 'Valid']
final_df2 = final_df[final_df.type == 'XSS']
final_df3 = final_df[final_df.type == 'SqlInjection']
final_df4 = final_df[final_df.type == 'PathTransversal']
final_df5 = final_df[final_df.type == 'OsCommanding']
new_final = pd.concat([final_df1, final_df2, final_df3, final_df4, final_df5])

# Visualize dataset composition
type_counts = new_final['type'].value_counts()
plt.figure(figsize=(10, 6))
bars = plt.bar(type_counts.index, type_counts.values, color=['#3498db', '#e74c3c', '#2ecc71', '#f39c12', '#9b59b6'])
plt.title('ECML Dataset Composition by Attack Type', fontsize=14, fontweight='bold', pad=20)
plt.xlabel('Attack Type', fontsize=12, fontweight='bold')
plt.ylabel('Number of Samples', fontsize=12, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3, linestyle='--')

# Add value labels on bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height):,}',
             ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()

# Display statistics
print(f"\nDataset Statistics:")
print(f"Total samples: {len(new_final):,}")
print(f"\nClass distribution:")
for attack_type, count in type_counts.items():
    percentage = (count / len(new_final)) * 100
    print(f"  {attack_type}: {count:,} ({percentage:.2f}%)") 
```
